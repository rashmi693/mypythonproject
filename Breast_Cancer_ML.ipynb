```python
x=df.iloc[:,2:31]
y=df.iloc[:,1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
x_test.head()
```

```python
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
x_train
```

```python
def models(x_train,y_train):
    from sklearn.linear_model import LogisticRegression
    log=LogisticRegression(random_state=0)
    log.fit(x_train,y_train)
    
    from sklearn.tree import DecisionTreeClassifier
    tree=DecisionTreeClassifier(criterion='entropy',random_state=0)
    tree.fit(x_train,y_train)
    
    from sklearn.ensemble import RandomForestClassifier
    forest=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
    forest.fit(x_train,y_train)
    
    print('Logistic Accuracy: ',log.score(x_test,y_test))
    print('DecesionTree Accuracy: ',tree.score(x_test,y_test))
    print('RandomForest Accuracy: ',forest.score(x_test,y_test))
    return log,tree,forest
```

```python
model=models(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
for i in range(len(model)):
    print('Model',i)
    cn=confusion_matrix(y_test,model[i].predict(x_test))
    TP=cn[0][0]
    TN=cn[1][1]
    FN=cn[1][0]
    FP=cn[0][1]
    print('Testing Accuracy =',(TP+TN)/(TP+TN+FP+FN))
     print('Recall = ',TP/(TP+FN))
    print('Precesion = ',TP/(TP+FP))
```
